```python
df=df.drop_duplicates().reset_index(drop=True)
```

```python
df=df[["title","overview",'genres',"tagline","vote_average","popularity"]]
```

```python
df= df.dropna(subset=["title"])
```

```python
df["overview"]=df["overview"].fillna('')
```

```python
df.iloc[0]['genres']
import ast
```

```python
df['genres']=df["genres"].apply(lambda x :" ".join([i["name"]for i in ast.literal_eval(x)]))
```

```python
df.head()
```

```python
df["tagline"]=df["tagline"].fillna('')
```

```python
df.isnull().sum()
```

```python
df['tags']=df['overview']+" "+df["genres"]+" "+ df["tagline"]
```

```python
df.head()
```

```python
df['tags'] = df['tags'].apply(lambda x : x.lower())
```

```python
import string

def remove_punc(txt):
  return txt.translate(str.maketrans('','',string.punctuation))
```

```python
df['tags'] = df['tags'].apply(remove_punc)
```

```python
def remove_numbers(txt):
    new = ""
    for i in txt:
        if not i.isdigit():
            new = new + i
    return new

df['tags'] = df['tags'].apply(remove_numbers)
     

def remove_emojis(txt):
    new = ""
    for i in txt:
        if i.isascii():
            new += i
    return new

df['tags'] = df['tags'].apply(remove_emojis)
```

```python
df["tags"][1]
```

```python
import nltk
from nltk.corpus import stopwords 
from nltk.stem import WordNetLemmatizer 
import re 
```

```python
nltk.download("stopwords")
nltk.download("wordnet")
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
    text = str(text).lower()
    #removed the puncuation 
    text=re.sub(r'[^a-zA-Z\s]',"",text)
    words=text.split()
    #remove ststopwords 
    words =[word for word in words if word not in stop_words]
    # lemmatize
    words=[lemmatizer.lemmatize(word) for word in words]
    return " ".join(words)
df['tags'] = df['tags'].apply(preprocess_text)
```

```python
df["tags"][1]
```

```python
df=df.reset_index(drop=True)
```

```python
indices = pd.Series(df.index,index = df["title"]).drop_duplicates()
indices
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features=50000,ngram_range=(1,2),stop_words="english")
```

```python
tfidf_matrix = tfidf.fit_transform(df["tags"])
```

```python
tfidf_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend(title,n=10): 
    if title not in indices : 
        return ["Movie not found"] 

    idx=indices[title]
    sim_score = cosine_similarity(tfidf_matrix[idx],tfidf_matrix).flatten()
    similar_idx = sim_score.argsort()[::-1][1:n+1] 
    return df["title"].iloc[similar_idx]
```

```python
recommend("Smoke")
```
